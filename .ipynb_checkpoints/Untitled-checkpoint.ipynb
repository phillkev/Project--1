```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Take in four years data and read it into pandas, the decimal and thousands things do not work
bls_2014 = pd.read_csv("bls_cve/bls_2014.csv",decimal=',')
bls_2015 = pd.read_csv("bls_cve/bls_2015.csv",thousands=',')
bls_2016 = pd.read_csv("bls_cve/bls_2016.csv",thousands=',')
bls_2017 = pd.read_csv("bls_cve/bls_2017.csv",thousands=',')

# Clean the data
# I need a little more explaination about the below line
bls_2014=bls_2014[bls_2014.TOT_EMP!='**']  
bls_2014.TOT_EMP=[float(i) for i in bls_2014.TOT_EMP]
bls_2014.EMP_PRSE=[float(i) for i in bls_2014.EMP_PRSE]
bls_2015=bls_2015[bls_2015.TOT_EMP!='**']  
bls_2015.TOT_EMP=[float(i) for i in bls_2015.TOT_EMP]
bls_2015.EMP_PRSE=[float(i) for i in bls_2015.EMP_PRSE]
bls_2016=bls_2016[bls_2016.TOT_EMP!='**']  
bls_2016.TOT_EMP=[float(i) for i in bls_2016.TOT_EMP]
bls_2016.EMP_PRSE=[float(i) for i in bls_2016.EMP_PRSE]
bls_2017=bls_2017[bls_2017.TOT_EMP!='**']  
bls_2017.TOT_EMP=[float(i) for i in bls_2017.TOT_EMP]
bls_2017.EMP_PRSE=[float(i) for i in bls_2017.EMP_PRSE]

bls_2015.NAICS=[i//10000 for i in bls_2015.NAICS]
bls_2016.NAICS=[i//10000 for i in bls_2016.NAICS]
# bls_2017.loc[bls_2017.NAICS=='3250A1','NAICS']='325000'
# bls_2017[bls_2017.NAICS=='3320A2']
# bls_2017.NAICS=[float(i)//10000 for i in bls_2017.NAICS]
bls_2014.NAICS=[i[0]+i[1] for i in bls_2014.NAICS]
# bls_2015.NAICS=[i[0]+i[1] for i in bls_2015.NAICS]
# bls_2016.NAICS=[i[0]+i[1] for i in bls_2016.NAICS]
bls_2017.NAICS=[i[0]+i[1] for i in bls_2017.NAICS]
bls_2017[bls_2017.NAICS=='72']
```

```python
# bls_g=[]
# bls_g.append(i.groupby('OCC_TITLE') for i in bls_try)
# bls_g[0].count()
# bls_ins_2014=bls_2014.groupby(['NAICS','NAICS_TITLE'])
# x['Other Financial Investment Activities']
# bls_ins_2014.count().sum().TOT_EMP
# bls_2014.TOT_EMP.sum()
bls_ins_2014=bls_2014.groupby(['NAICS'])
bls_ins_2015=bls_2015.groupby(['NAICS'])
bls_ins_2016=bls_2016.groupby(['NAICS'])
bls_ins_2017=bls_2017.groupby(['NAICS'])
# bls_ins_2014.sum().sort_values('TOT_EMP',ascending=False).head(30).TOT_EMP
df2014=bls_ins_2014.sum().TOT_EMP
df2015=bls_ins_2015.sum().TOT_EMP
df2016=bls_ins_2016.sum().TOT_EMP
df2017=bls_ins_2017.sum().TOT_EMP
df=pd.DataFrame({'2014':df2014,'2015':df2015,'2016':df2016,'2017':df2017})
ddf=pd.DataFrame({'Natural Resources and Mining 1':df.iloc[0],'Construction 2':df.iloc[1]+df.iloc[2]+df.iloc[3],'Manufacturing 3':df.iloc[4]+df.iloc[5]+df.iloc[6],
'Trade, Transportation, and Utilities 4':df.iloc[7]+df.iloc[8]+df.iloc[9]+df.iloc[10]+df.iloc[11],
'Information 51':df.iloc[12], 'Financial Activities 52 53':df.iloc[13]+df.iloc[14],'Professional and Business Services 54 55 56':df.iloc[15]+df.iloc[16]+df.iloc[17], 
'Education and Health Services 6':df.iloc[18]+df.iloc[19],
'Leisure and Hospitality 7':df.iloc[20]+df.iloc[21],'Social Organizations 8':df.iloc[22],'Government 9':df.iloc[23]})
# df.iloc[7]+df.iloc[8]+df.iloc[9]+df.iloc[10]+df.iloc[11]
ddf
```
